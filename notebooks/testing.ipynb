```python
initial_data = clean.read_file('../data/master.csv')
initial_data
```

```python
data = clean.clean_data(initial_data, 2016, '5-14 years',
                            '05-14 years', ['country-year', 'HDI for year'])
```

```python
v_five = data[(data.Country == 'Russian Federation') | (data.Country == 'United States') | (data.Country == 'Japan') | (data.Country == 'France') | (data.Country == 'Ukraine')]
v_five = v_five.sort_values('Suicides', ascending=False)

fig, ax = plt.subplots(figsize=(9,6))

sns.boxplot(x="Country", y="Suicides", data=v_five, ax=ax)
ax.set_title('High Five Distribution', loc='left', fontsize=18, fontweight='bold')

ax.set_xlabel('Country', fontsize=12, fontweight='medium')
ax.set_ylabel('Suicides', fontsize=12, fontweight='medium')
```
